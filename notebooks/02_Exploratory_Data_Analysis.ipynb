# Step 2 - Exploratory Data Analysis

The Exploratory Data Analysis (EDA) phase helps us gain a deeper understanding of the datasetâ€™s structure and characteristics. Through EDA, we uncover patterns, identify anomalies, and extract valuable insights that guide subsequent modeling steps. 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## 2.0 - Load the Data

```python
# Load the CSV file for the previous data
previous_years_df = pd.read_csv('../data/02_previous_years_not_null.csv')
print(f"The shape of the dataset is: {previous_years_df.shape}")

# Double check for missing values
print(f"Are missing values?: {previous_years_df.isnull().values.any()}")
```

```python
# Load the CSV file for the present data
present_year_df = pd.read_csv('../data/04_present_year_not_null.csv')
print(f"The shape of the dataset is: {present_year_df.shape}")

# Double check for missing values
print(f"Are missing values?: {present_year_df.isnull().values.any()}")
```

## 2.1 - Class Imbalance Analysis

Previous Years

```python
# Calculate the counts for each class in both dataframes
previous_counts = previous_years_df['class'].value_counts()
previous_counts
```

```python
# Plot
plt.style.use("dark_background")
fig, ax = plt.subplots(figsize=(6, 4))
tcb = previous_counts.plot(kind='bar', ax=ax, color=sns.color_palette("rocket"))
ax.set_title('Defect Type Distribution - Previous Years')
ax.set_xlabel('Class')
ax.set_ylabel('Number of Classes in the Dataset')
plt.xticks(rotation=0)
plt.grid(axis='y', alpha=0.7)
plt.show()
```

Present Year

```python
# Calculate the counts for each class in both dataframes
present_counts = present_year_df['class'].value_counts()
present_counts
```

```python
# Plot
plt.style.use("dark_background")
fig, ax = plt.subplots(figsize=(6, 4))
tcb = present_counts.plot(kind='bar', ax=ax, color=sns.color_palette("rocket"))
ax.set_title('Defect Type Distribution - Present Year')
ax.set_xlabel('Class')
ax.set_ylabel('Number of Classes in the Dataset')
plt.xticks(rotation=0)
plt.grid(axis='y', alpha=0.7)
plt.show()
```

The class imbalance in the dataset is quite significant. 

In this case, we can apply some resampling techniques, like:

**Oversampling:** Increase the number of instances in the minority class by duplicating existing ones or generating synthetic samples (e.g., using SMOTE - Synthetic Minority Over-sampling Technique).

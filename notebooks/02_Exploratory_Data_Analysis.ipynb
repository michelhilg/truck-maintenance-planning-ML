```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Load the CSV file for the present data
present_year_df = pd.read_csv('../data/air_system_present_year.csv')
present_year_df
```

```python
# Load the CSV file for the previous data
previous_years_df = pd.read_csv('../data/air_system_previous_years.csv')
previous_years_df
```

```python
# Basic information about the datasets
print(previous_years_df.info())
print(present_year_df.info())

# Summary statistics
print(previous_years_df.describe())
print(present_year_df.describe())
```

```python
print(previous_years_df.isna().sum())
```

```python
# List columns
print(previous_years_df.columns)
print(present_year_df.columns)

# Check for common columns
common_columns = set(previous_years_df.columns).intersection(set(present_year_df.columns))
print(f"Common Columns: {common_columns}")
```

```python
# Replace 'na' with np.nan and then drop rows with any NaN values
previous_years_df.replace('na', pd.NA, inplace=True)
present_year_df.replace('na', pd.NA, inplace=True)

# Check for missing values and print only columns with missing values
missing_previous = previous_years_df.isna().sum()
missing_present = present_year_df.isna().sum()

print("Columns with missing values in previous_years_df:")
print(missing_previous.sort_values(ascending=False))

print("\nColumns with missing values in present_year_df:")
print(missing_present.sort_values(ascending=False))
```

```python


# Drop the 'class' column
previous_years_df = previous_years_df.drop(columns=['class'])
present_year_df = present_year_df.drop(columns=['class'])

# Correlation matrices
previous_corr = previous_years_df.corr()
present_corr = present_year_df.corr()

# Visualize the correlation matrices
plt.figure(figsize=(12, 8))
sns.heatmap(previous_corr, cmap='coolwarm')
plt.title('Correlation Matrix for Previous Years Data')
plt.show()

plt.figure(figsize=(12, 8))
sns.heatmap(present_corr, cmap='coolwarm')
plt.title('Correlation Matrix for Present Year Data')
plt.show()
```

### 2 - What is the percentage of trucks with defects? (1pt)

```python
# Concat the dataframes
combined_df = pd.concat([previous_years_df, present_year_df])

# Count the occurrences of 'pos' and 'neg'
pos_count = combined_df['class'].value_counts().get('pos', 0)
neg_count = combined_df['class'].value_counts().get('neg', 0)
total_count = pos_count + neg_count

# Calculate percentages
if total_count > 0:
    pos_percentage = (pos_count / total_count) * 100
    neg_percentage = (neg_count / total_count) * 100
else:
    pos_percentage = neg_percentage = 0

print(f"Percentage of trucks with air system defects (pos): {pos_percentage:.2f}%")
print(f"Percentage of trucks with other defects (neg): {neg_percentage:.2f}%")
```
